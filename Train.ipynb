```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset
import torch

# Dataset yükleme
dataset = load_dataset("anilguven/turkish_news_dataset")

# Tokenizer ve veri işleme 
tokenizer = AutoTokenizer.from_pretrained("dbmdz/bert-base-turkish-cased")
max_length = 512

def preprocess_function(examples):
    return tokenizer(
        examples['HABERLER'],  
        padding="max_length",
        truncation=True,
        max_length=max_length
    )


encoded_dataset = dataset.map(preprocess_function, batched=True)

# Label sütunlarını integer'a çevirme 
label_column = "HABERLER"
label_mapping = {label: i for i, label in enumerate(set(dataset["train"][label_column]))}

def add_labels(examples):
    examples["labels"] = [label_mapping[label] for label in examples[label_column]]
    return examples

encoded_dataset = encoded_dataset.map(add_labels, batched=True)

# Dataset'i PyTorch formatına dönüştürme
encoded_dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)

# Model yükleme
model = AutoModelForSequenceClassification.from_pretrained(
    "dbmdz/bert-base-turkish-cased",
    num_labels=len(label_mapping)  
)

# Eğitim ayarları
training_args = TrainingArguments(
    fp16=True,  # Daha hızlı eğitim için 
    gradient_accumulation_steps=2,  
    output_dir="./results",  
    logging_dir="./logs",  
    evaluation_strategy="epoch",  # Her epoch'ta değerlendirme yapmak için kullanılır
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_steps=10,  
)

# Train ve val datasetlerini ayırma
eval_dataset = encoded_dataset["validation"] if "validation" in encoded_dataset else encoded_dataset["train"]

# Trainer 
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=encoded_dataset["train"],
    eval_dataset=eval_dataset,
)

# Eğitim
trainer.train()

# Model ve tokenizer'ı kaydetme
save_dir = "./trained_model"
model.save_pretrained(save_dir)
tokenizer.save_pretrained(save_dir)

print(f"Model ve tokenizer {save_dir} klasörüne kaydedildi.")
```
